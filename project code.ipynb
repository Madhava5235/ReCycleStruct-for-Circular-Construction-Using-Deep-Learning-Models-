```python
import tensorflow as tf

# Load the trained classifier
classifier = tf.keras.models.load_model("/content/drive/MyDrive/material_classifier.h5")

IMG_HEIGHT, IMG_WIDTH = 224, 224

# Function to classify a cropped object
def classify_object(image):
    img = image.resize((IMG_HEIGHT, IMG_WIDTH))
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    predictions = classifier.predict(img_array)
    class_names = ["Burnt Clay Brick - First Grade", "Burnt Clay Brick - Second Grade", "Burnt Clay Brick - Third Grade",
                   "Solid Concrete Block - First Grade", "Solid Concrete Block - Second Grade", "Solid Concrete Block - Third Grade"]

    return class_names[np.argmax(predictions)]
```

```python
# Detect and classify objects
def detect_and_classify(image_path):
    detected_objects, results = detect_objects(image_path)
    image = Image.open(image_path)

    classified_results = []
    for obj in detected_objects:
        x1, y1, x2, y2, label = obj

        # Crop detected object and classify
        cropped_object = image.crop((x1, y1, x2, y2))
        classification = classify_object(cropped_object)

        classified_results.append((label, classification))

    return classified_results

# Test on an image
image_path = "/content/drive/MyDrive/Bricks/Burnt_Clay_Brick_First_Grade/IMG-20250131-WA0031.jpg"
results = detect_and_classify(image_path)
print("Final Results:", results)
```

```python
import os

# Define dataset paths
dataset_path = "/content/drive/MyDrive/Bricks/"
labels_path = "/content/drive/MyDrive/Bricks/labels/"

# Mapping class names to YOLO class IDs
class_mapping = {
    "Burnt_Clay_Brick_First_Grade": 0,
    "Burnt_Clay_Brick_Second_Grade": 1,
    "Burnt_Clay_Brick_Third_Grade": 2,
    "Solid_Concrete_Block_First_Grade": 3,
    "Solid_Concrete_Block_Second_Grade": 4,
    "Solid_Concrete_Block_Third_Grade": 5
}

# Ensure the labels directory exists
os.makedirs(labels_path, exist_ok=True)

# Iterate through class folders and generate YOLO annotations
for class_name, class_id in class_mapping.items():
    class_dir = os.path.join(dataset_path, class_name)
    label_dir = os.path.join(labels_path, class_name)
    os.makedirs(label_dir, exist_ok=True)

    for img_name in os.listdir(class_dir):
        if img_name.endswith((".jpg", ".png", ".jpeg")):
            # Create label file with the same name as the image but with .txt extension
            label_file = os.path.join(label_dir, img_name.rsplit(".", 1)[0] + ".txt")

            # YOLO format: <class_id> <x_center> <y_center> <width> <height>
            # Since we don't have exact bounding boxes, we'll assume the entire image contains the object
            with open(label_file, "w") as f:
                f.write(f"{class_id} 0.5 0.5 1.0 1.0\n")  # Entire image is the object
```

```python
import shutil
import os

# Define paths
dataset_path = "/content/drive/MyDrive/Bricks/images/"
labels_path = "/content/drive/MyDrive/Bricks/labels/"

# Create 'train' and 'val' directories
for split in ["train", "val"]:
    os.makedirs(os.path.join(dataset_path, split), exist_ok=True)
    os.makedirs(os.path.join(labels_path, split), exist_ok=True)

# Copy all images and labels into both 'train' and 'val'
for class_name in os.listdir(dataset_path):
    class_path = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_path):
        for split in ["train", "val"]:
            dest_path = os.path.join(dataset_path, split, class_name)
            os.makedirs(dest_path, exist_ok=True)

            # Copy images
            for file in os.listdir(class_path):
                if file.endswith((".jpg", ".png", ".jpeg")):
                    shutil.copy(os.path.join(class_path, file), os.path.join(dest_path, file))

# Copy labels similarly
for class_name in os.listdir(labels_path):
    class_path = os.path.join(labels_path, class_name)

    if os.path.isdir(class_path):
        for split in ["train", "val"]:
            dest_path = os.path.join(labels_path, split, class_name)
            os.makedirs(dest_path, exist_ok=True)

            # Copy label files
            for file in os.listdir(class_path):
                if file.endswith(".txt"):
                    shutil.copy(os.path.join(class_path, file), os.path.join(dest_path, file))

print("✅ All data is now copied into both 'train' and 'val'.")
```
